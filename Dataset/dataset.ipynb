```python
def print_structure(d,indent=0):
  if isinstance(d,dict):
    for key,value in d.items():
      print('\t'*indent+str(key))
      print_structure(value,indent+1)

  if isinstance(d,list):
    print('\t'*indent + "[List of length {} containing:]".format(len(d)))
    if d:
      print_structure(d[0],indent+1)
```

```python
import json

def get_json_data(json_path):
  with open(json_path) as f:
    data = json.load(f)
  return data
```

```python
train_json_path = "C:\\Users\\Debajyoti\\OneDrive\\Desktop\\Brain tumor disease\\Dataset\\Brain tumor segmentation\\train\\_annotations.coco.json"
test_json_path = "C:\\Users\\Debajyoti\\OneDrive\\Desktop\\Brain tumor disease\\Dataset\\Brain tumor segmentation\\test\\_annotations.coco.json"

print("Fetching.... JSON 1")
print("Structure of Train JSON data")
print_structure(get_json_data(train_json_path))
print("Fetching.... JSON 2")
print("Structure of Test JSON data")
print_structure(get_json_data(test_json_path))
```
